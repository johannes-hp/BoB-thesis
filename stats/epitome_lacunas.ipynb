```python
import os
import re
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pattern = '\d+'

all_epitomes = list(range(140, 2331))
epitome_numbers = [int(re.search(pattern, file).group()) for file in os.listdir('../epitomes')]
```

```python
epitome_bools = [1 if epitome in epitome_numbers else 0 for epitome in all_epitomes]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(2, 6)
plt.tight_layout()
sns.heatmap(pd.DataFrame(epitome_bools, index=all_epitomes, columns=['']),
            cbar=False,
            cmap='Reds',
            ax=ax,)
ax.set_ylabel('Epitome no.')
ax.set_title('Overview of transcribed epitomes\n');
```
