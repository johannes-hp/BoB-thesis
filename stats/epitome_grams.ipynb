```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
epitomes_s = ''
for file in os.listdir('../epitomes'):
    with open(os.path.join('../epitomes', file), 'r', encoding='utf-8') as f:
        text = f.readlines()
        epitomes_s += text[-1]
```

```python
char_counter = Counter(epitomes_s.lower())
```

```python
most_common_letters = dict(char_counter.most_common(23)[1:])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
# seaborn is hilariously unwilling to work with anything other than lists when doing barplots
sns.barplot(x=list(range(len(most_common_letters.keys()))), 
            y=list(most_common_letters.values()),
            color='firebrick',
            ax=ax)

ax.set_xticklabels(most_common_letters.keys())
ax.set_ylabel('Frequency')
ax.set_title('The most common letters in the BoB');
```

```python
epitomes_tokens = epitomes_s.lower().split()
token_counter = Counter(epitomes_tokens)
most_common_tokens = dict(token_counter.most_common(30))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
# seaborn is hilariously unwilling to work with anything other than lists when doing barplots
sns.barplot(x=list(range(len(most_common_tokens.keys()))), 
            y=list(most_common_tokens.values()),
            color='firebrick',
            ax=ax)

ax.set_xticklabels(most_common_tokens.keys(), rotation=75)
ax.set_ylabel('Frequency')
ax.set_title('The most common tokens in the BoB');
```

```python
token_bigrams = [f'{epitomes_tokens[i]} {epitomes_tokens[i+1]}' for i in range(len(epitomes_tokens)-1)]
token_bigrams_counter = Counter(token_bigrams)
most_common_token_bigrams = dict(token_bigrams_counter.most_common(25))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
# seaborn is hilariously unwilling to work with anything other than lists when doing barplots
sns.barplot(x=list(range(len(most_common_token_bigrams.keys()))), 
            y=list(most_common_token_bigrams.values()),
            color='firebrick',
            ax=ax)

ax.set_xticklabels(most_common_token_bigrams.keys(), rotation=75)
ax.set_ylabel('Frequency')
ax.set_title('The most common token bigrams in the BoB');
```

```python
char_bigrams = [f'{epitomes_s[i]}{epitomes_s[i+1]}' for i in range(len(epitomes_s)-1)]
char_bigrams_counter = Counter(char_bigrams)
most_common_char_bigrams = dict(char_bigrams_counter.most_common(25))
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8, 6)
# seaborn is hilariously unwilling to work with anything other than lists when doing barplots
sns.barplot(x=list(range(len(most_common_char_bigrams.keys()))), 
            y=list(most_common_char_bigrams.values()),
            color='firebrick',
            ax=ax)

ax.set_xticklabels(most_common_char_bigrams.keys())
ax.set_ylabel('Frequency')
ax.set_title('The most common character bigrams in the BoB');
```
